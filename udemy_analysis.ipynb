```python
del_index = df[(df["course_hours"] == "41totalmins") | (df["course_hours"] == "43totalmins")
   | (df["course_hours"] == "37totalmins") | (df["course_hours"] == "31totalmins")
    | (df["course_hours"] == "32totalmins") | (df["course_hours"] == "38totalmins")
    | (df["course_hours"] == "44totalmins") | (df["course_hours"] == "34totalmins")
    | (df["course_hours"] == "42totalmins") | (df["course_hours"] == "35totalmins")
    | (df["course_hours"] == "40totalmins") | (df["course_hours"] == "39totalmins")
  | (df["course_hours"] == "30totalmins") | (df["course_hours"] == "36totalmins")
  | (df["course_hours"] == "33totalmins")].index
```

```python
df = df.drop(del_index, axis = 0)
df.info()
```

```python
df.avg_rating = df.avg_rating.astype("float64")
df.reviews_count = df.reviews_count.astype("int64")
df.course_hours = df.course_hours.astype("float64")
df.lectures_count = df.lectures_count.astype("int64")
df.price_after_discount = df.price_after_discount.astype("int64")
df['main_price'] = df['main_price'].astype("int64")
df['student_count'] = df['student_count'].astype("int64")
df.head()
```

```python
from scipy.stats import linregress
x=df['student_count']
y=df['reviews_count']
slope, intercept, r_value, p_value, std_err = linregress(x,y )
x_range = [min(x), max(x)]
y_reg = [slope * x + intercept for x in x_range]
plt.plot(x, y, '|', label='Data points')
plt.plot(x_range, y_reg, label='Regression line')
plt.xlabel('count of students enrolled the course')
plt.ylabel('count of reviews recieved for the course')
plt.title('Linear Regression')
plt.ylim(y.quantile(0.01),y.quantile(0.99))
plt.xlim(x.quantile(0.01),x.quantile(0.99))
plt.legend()
plt.show()
equation = f"y = {slope:.2f}x + {intercept:.2f}"
print("Equation:", equation)
```
